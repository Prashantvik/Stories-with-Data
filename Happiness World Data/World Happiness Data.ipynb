## Story 2 : World Happiness Data

The World Happiness Report is a landmark survey of the state of global happiness that ranks 156 countries by how happy their citizens perceive themselves to be. The World Happiness Report 2020 for the first time ranks cities around the world by their subjective well-being and digs more deeply into how the social, urban and natural environments combine to affect our happiness.

The World Happiness Report is a publication of the Sustainable Development Solutions Network, powered by data from the Gallup World Poll, and supported by the Ernesto Illy Foundation, illycaffè, Davines Group, Blue Chip Foundation, the William, Jeff, and Jennifer Gross Family Foundation, and Unilever’s largest ice cream brand Wall’s.

```python
#Importing the required libraries for analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

plt.rcParams["figure.figsize"] = 8,5
```

```python
#Reading the Hapiness data by Country of year 2015
df = pd.read_csv("2015.csv")

#Reading the Hapiness data by Country of year 2016 and 2017
df16 = pd.read_csv("2016.csv")
df17 = pd.read_csv("2017.csv")
```

```python
#Features (Columns) in the dataset
df.head(3)
```

What is Dystopia?
Dystopia is an imaginary country that has the world’s least-happy people. The purpose in establishing Dystopia is to have a benchmark against which all countries can be favorably compared (no country performs more poorly than Dystopia) in terms of each of the six key variables, thus allowing each sub-bar to be of positive (or zero, in six instances) width. The lowest scores observed for the six key variables, therefore, characterize Dystopia. Since life would be very unpleasant in a country with the world’s lowest incomes, lowest life expectancy, lowest generosity, most corruption, least freedom, and least social support, it is referred to as “Dystopia,” in contrast to Utopia.

## Visualization and Insights!

#### We will mostly be using Happiness Score rather than Happiness just for the sake of better interpretation and unsertanding, since rank will be low for some parameter that increases the country's happiness. 

```python
plt.plot("Happiness Score","Economy (GDP per Capita)", data = df, color="darkorange")
plt.xlabel("Hapiness Score")
plt.ylabel("GDP")
plt.title("Happiness Rank vs GDP graph")           
```

We can see the most (Not all) of the countries follow the trend of higher the GDP of the country, higher the Happiness score which leads to high ranking of the country in Happiness list.This certainly is a big factory, people of a country which has high GDP anc low poverty are more happy when compairing to the other countries. There's a correlation between happiness and wealth, to a point of about $75,000 per year.

When people make more than 75k dollar a year, their happiness doesn't increase, but the lower their income is the worse they feel, the study found.

Along with GDP, let's consider some other factors too and visualize the relationship and how other factors of a country affects the Happiness of a country.

```python
plt.plot("Family","Happiness Rank", data = df, color="crimson")
plt.ylabel("Hapiness Score")
plt.xlabel("Family")
plt.title("Happiness Rank vs Family Index")  
```

Data shows thet as the family index increases the Happiness Score decreases.

Having large family(High family index) could affect the Happiness of a family, hence the Happiness Rank of that country. 

```python
#Similar visulaization with other features!
plt.plot("Happiness Rank","Freedom", data = df)
plt.xlabel("Hapiness Rank")
plt.ylabel("Freedom")
plt.title("Happiness Rank vs Freedom Index")
```

We can interpret here, there is a link between freedom and happiness: “One of the reasons for the political pursuit of freedom is the belief that this will add to greater happiness for a greater number of citizens."

```python
sns.regplot("Freedom", "Happiness Score", data = df, color ="lime")
plt.title("Relationship of Freedom with Happiness Score")
```

"The common-sense theory behind this belief is that life will be more satisfying if we can live the way we want.”

If we plot the Happiness Score with Freedon on the mathematical grounds, we will find out the direct relationship between both of them. Regression plot above show the linear relation.

```python
plt.plot("Happiness Score","Health (Life Expectancy)", data = df, color="darkviolet")
plt.xlabel("Hapiness Score")
plt.ylabel("Health (Life Expectancy)")
plt.title("Health (Life Expectancy) vs Happiness Score")
```

Happy life expectancy is longer than active life expectancy. And there has been greater compression of unhappiness than compression of morbidity. There are substantial differentials in happy life expectancy by sex and race because of differential prevalence rates of happiness.

### Region wise Analysis

```python
#Happiness Score Region wise
sns.boxplot(x = "Region", y = "Happiness Score", data = df.sort_values(by="Happiness Score", ascending = False))
plt.xticks(rotation=90)
plt.title("Boxplots")
```

As per the boxplot, we can see the Region of Western europe have the higest of median followed by North America. If we consider Southern Asia, it has median of about 4.5 and have position of about second last in all the regions. 

```python
#Defining the new dataframe and group it by 'Region'
Grouped = df.groupby("Region").mean()
```

```python
Grouped_sort = Grouped.sort_values(by = "Happiness Score", ascending = False)
plt.plot("Happiness Score", data = Grouped_sort , color="orchid")
plt.xticks(rotation = 90)
plt.xlabel("Region")
plt.ylabel("Happiness Score")
plt.title("Happiness Score Mean by Region")
```

Happiness Score mean of Australia and New Zeland Region of all the Regions, followed by North America while Southern Asia stands at the second last position. Whereas Eastern Asia stand on the middle, the peak at the Southeastern Asia has high Happiness Score mean than Middle East and Northern Africa, Aouthern Asia & Middle East and Northern Europe.
